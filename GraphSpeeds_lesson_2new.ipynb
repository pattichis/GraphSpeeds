# Tables and object-oriented programming
In coding, objects are used to save functions and their data for processing.
We can create an object using a <tt>constructor</tt>.

Here we create a table object to store information using:
```
Elephant = table()
```
We can then store data using:
```
Elephant.column_labels = ["x (hours)", "y (miles)"]
time_values     = [0, 1, 2, 3, 4, 5]
distance_values = [0, 1, 2, 3, 4, 5] # <- Speed=1 mile/hour
Elephant.data_values = [time_values, distance_values]
```
We can see what we stored using <tt>print()</tt>.

However, a nicer way to see the values is to use <tt>plotTable():</tt>
```
Elephant.plotTable()
```
You can save an image using <tt>saveImage()</tt> as shown below:
```
# Create and plot the tyler table first ...
Elephant.saveImage("Elephant-table.png")
```
This will save the table in the file <tt>tyler.png</tt>.

## Place your characters on an image
In order to build the video, you need to understand the size of your image:
```
 600 pixels height means you have 600 rows from the top to the bottom.
 800 pixels width  means you have 800 columns from left to right.
```

You thus want to allocate a number of pixels so that you can put multiple pictures above each other.

The recommendation is to use:
```
target_size = 50
```
This means that each character will take 50 pixels out of the 600 pixels for moving up and down.

The next thing is to understand how to place your chacters!
This is done by specifying the row and column.<br>
```
Elephant.loc = (0, target_size)   # 0 is the leftmost column
                                  # target_size is the bottom row of the character
```

# Make videos VII: Make the final video
You can make the final video by combining images and videos together.
```
img_list = ["multi-line.png",  "plot.jpg", "video.mp4", "tyler.jpg"]
```
Here, note the following:
- multi-line.png is an image. The file name ends in .png
- plot.jpg       is an image. The file name ends in .jpg
- video.mp4      is a video. The file name ends in .mp4
- tyler.jpg      is an image. The file name ends in .jpg

For the images, we can specify how long to display each one.<br>
Suppose that we want:
- 2.0 seconds for multi-line.png.
- 1.0 seconds for plot.jpg.
- 1.0         for video.mp4. This number is IGNORED.      
- 1.0 seconds for tyler.jpg.
We put all of these numbers in a list as given by:
```
durations_list = [2, 1.0, 1.0, 1.0]
```

We can now create the video using:
```
video=CreateVideo(video_name="all.mp4",       # This is file name of the final video.  
                  file_list=img_list,         # This the list of images and files to combine.
                  fps=10,                     # The number of frames per second.
                  durations=durations_list)   # The duration for each one.
```
We can then display the video using:
```
video.ipython_display()
```
## Assignment: Make final video
Follow the process to create your final video:
1. Determine which images and videos you want to use.
2. Determine the order. For example, what is displayed first, second, etc.
3. Create the <tt>img_list</tt> array with the file names.
4. Determine how many seconds you want to display each image.
5. Create the <tt>durations_list</tt> array to enter durations for each image (enter 1 for videos).
6. Determine the final video file name.
7. Determine the frames per second. Here, note that the fps should match the race video.
8. Run <tt>CreateVideo()</tt> and <tt>video.ipython_display()</tt> to debug your video.


````python
# Decide which images and videos you want to put together:
img_list = ["multi-line.png",  "plot.jpg", "race.mp4", "Single-line.png"]

# Set the duration for each one in seconds. Note: video durations are ignored.
durations_list = [2, 1.0, 1.0, 1.0]

# Create the video:
video=CreateVideo(video_name="all.mp4",       # This is file name of the final video.
                  file_list=img_list,         # This the list of images and files to combine.
                  fps=10,                     # The number of frames per second.
                  durations=durations_list)   # The duration for each one.

# Display the video:
video.ipython_display()
````
